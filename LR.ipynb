```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
excel_file_path = 'house_price.xlsx'
df = pd.read_excel(excel_file_path)
df.head()
```

```python
%matplotlib inline
plt.xlabel('area(sqr ft)')
plt.ylabel('price(US$)')
plt.scatter(df.area, df.price, color='red', marker="+")
```

```python
reg = linear_model.LinearRegression()  # creating an object of Linear regression
reg.fit(df[['area']], df.price)  # reg.fit is mainly training the linear regression model
```

After training the model let us predict some prices of house based on there sqr feet

```python
# import numpy as np # import numpy for reshaping

reg.predict(np.array(3300).reshape(-1, 1))
```
